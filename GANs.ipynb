# Generative Adversarial Networks

## Usage
GANs are used for generating realistic data.

Examples:
* STACKGAN MODEL: text description -> highres photo matching that description
    * Google creating image captions
* IGAN (Pix2Pix): user draws crude sketches -> searches and returns nearest possible realistic image
    * Adobe drawing/ sketch program
    * (domain #1 -> domain #2) blueprints of a building -> photos of finished buildings
* CycleGAN: (Berkley) usupervised real-time translation of images/ video

Facebook research: image -> image models can be trained in an unsupervised way
   * example generating avatars from real profile pics

Also used to create realistic simulated training sets and environments to train other models or RL agents.
   * (Apple research) on making eyes more realistic
   
Used for imitation learning:
   * learn actions (data) from an expert

Used to create adverserial examples:
   * fooling ML classifiers > security 

Used for **semi-supervised learning**



## How?
RNNs can be used to genarate sentences, one word at a time. And images, one pixel at a time (Autoregressive Models / Fully Visible Belief Networks).

GANs are a type of generative model that let's us generative an entire image in one shot.
### Generator Network
* differentiable function (neural network)
* takes random noise as input
* transforms and reshapes the input recognisable structure
* outputs a realistic image



The training process for a GAN is different from a supervised learning model. 
### Discriminator Network
* guides the generator
* regular neural network classifier
* during training the discriminator is shown real data and fake data (generated)
* the discriminator outputs the prob whether the input is real or fake.

Meanwhile the generator is trained to do the opposite -> it's trying to fool the discriminator to belief that it's output is real and therefor classified as real.



# Games and Equilibria

**Adversarial** means that the two networks (generator and discriminator) are in competition with each other -> Game Theory

### Game Theory
* model cooporation and conflict between rational agents
* Rock, Paper, Scissor example
    * randomize moves -> **equilibrium** -> player choses action uniformly at random
* Equilibrium: neither player can improve their situation without changing the other player's strategy
    * maximum for the descriminator and minimum for the generator (saddle-point)
