# Hyperparameters

### 2 types of hyperparameters:

* **Optimizer Hyperparameters**: variables related to the optimizer and training
    * `learning_rate`,
    * `batch_size` (minibatch size),
    * `epochs`

* **Model Hyperparameters**: variables related to structure of the model
    * `n_layers` and `n_units`

### Learning Rate

***The single most important hyperparameter and one should always make sure that it has been tuned***

* Good starting point = 0.01
* usual suspects: `0.1, 0.01, 0.001, 0.0001, 0.00001, 0.000001`
* learning rate decay
* adaptive learning (adam, adagrad, ...)

![](learning_rate.png)

NOTE: The above guidelines are for very simple models and practice is much more comblicated.

### (Mini)batch size
* impacts memory and hardware requirements
* usual suspects: 1,2,4,8,16, **32,64,128,256**
* a larger batch size allows computational boosts for matrix computations
* a smaller batch size improves the general learning (longer learning time)


### Epochs

* train as long as the error rate decreases
* early stopping: stop training with error stops decreasing
* Tensorflow and other DL libraries has multiple tools for early stopping


### Number of Hidden Units / Layers


Usually in deep learning, the more layers thus better. Big, is usually not a problem with techniques such as dropout, L2, etc. But way to big can be an issue. 

* as Jeremy Horward said, start by trying to overfit, then add regularization, dropout etc. Afterwards change layers if needed.

## RNN Hyperparameters

* Choosing a cell type: LSTMs vs. GRUs
* Depth of the model: # layers
* Embedding dimensionality: if working with text (usually 50-200, 500 or even 1000)
